```python
# convert this into a function
```

```python
def table_conversion(index):
    r=[]
    heading=[head.text.strip() for head in index.find_all('th')]
    for row in index.find_all('tr'):
        rows=[data.text.strip() for data in row.find_all('td')]
        r.append(rows)
    return [heading,r]
```

```python
a,b=table_conversion(table_tag[1])
```
